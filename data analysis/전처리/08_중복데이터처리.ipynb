#### 결측치, 이상치 처리 복습

```python
# bicycle_dup.csv 파일 읽어들인 후 조회
file_path = './data/bicycle_dup.csv'
df = pd.read_csv(file_path, encoding='euc-kr')
df
```

```python
# 결측치 확인
df.info()
```

```python
# 이상치 확인
df.describe()
```

```python
# df_temp 데이터를 이용해서 결측치 및 이상치 데이터 처리 연습
# df는 사용하지 않습니다. 처리 방법은 자유롭게 진행하세요
df_temp = df.copy()
df_temp
```

```python
df_temp = df.fillna(df_temp.mean())
df_temp
```

```python
df_temp.info()
```

```python
import numpy as np
q1, q3 = np.percentile(df_temp['나이'], [25,75])
print(q1)
print(q3)

IQR = q3 -q1

upper = q3 + (1.5*IQR)
lower = q1 - (1.5*IQR)
```

```python
(df_temp['나이'] > upper) | (df_temp['나이'] < lower)
```

```python
def outlier(df, data) :
    q1, q3 = np.percentile(data, [25,75])
    IQR = q3 -q1
    upper = q3 + (1.5*IQR)
    lower = q1 - (1.5*IQR)
    
    outlier = df[(data > upper) | (data < lower)]
    return outlier
```

```python
outlier(df_temp, df_temp['나이'])
```

```python
outlier(df_temp, df_temp['나이']).index
```

```python
df_temp = df_temp.drop(outlier(df_temp, df_temp['나이']).index, axis=0)
df_temp.reset_index(inplace=True, drop=True)
df_temp
```
