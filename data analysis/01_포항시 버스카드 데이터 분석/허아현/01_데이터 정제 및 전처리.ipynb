## 데이터 정제 및 전처리

```python
# 데이터 파일 불러오기
file_path1 = 'C:/STUDY/01_포항시 버스카드 데이터 분석/data/trfcard(1)'
df_ex = pd.read_csv(file_path1)
df_ex
```

```python
# 데이터정의서 파일 불러오기
file_path2 = './data/trfcard(0)/trfcard_columns.xlsx'
df_col_ex = pd.read_excel(file_path2, skiprows=[0,1], usecols='B:C')
df_col_ex
```

```python
# 컬럼명 변경
df_kor_ex = pd.DataFrame()
n_dict = {}

for i in range(len(df_col_ex)) :
    n_dict[df_ex.columns[i]] = df_col_ex['컬럼명 (한글)'][i]
    
df_kor_ex = df_ex.rename(columns = n_dict)
df_kor_ex
```

```python
# 데이터 합치기
df_e = pd.DataFrame()

for i in range(0, 80) :
    bus = pd.read_csv('./data/trfcard({})/trfcard.csv'.format(i))
    df_e = pd.concat([df_e, bus], ignore_index=True)

# 컬럼명 변경
df = pd.DataFrame()
n_dict = {}

df_col = pd.read_excel('./data/trfcard(0)/trfcard_columns.xlsx', skiprows=[0,1], usecols='B:C')

for j in range(len(df_col)) :
    n_dict[df_e.columns[j]] = df_col['컬럼명 (한글)'][j]

df = df_e.rename(columns = n_dict)
df
```

```python
# 결측치, 이상치 확인
df.info()
df.describe()
```

```python
# 중복데이터 확인
df[df.duplicated()]
```

```python
# 알수없는 값을 확인함(None)
df['승객연령'].unique()
```

```python
'''
교통카드 연령 구분을 구글링해본 결과 : 1회: 일반, 2회: 청소년·어린이, 3회: 에러
None은 잘 못 찍은 경우 다시 카드를 찍은 것으로 확인
제거해도 분석에 지장이 없을 것으로 판단되어 제거하고 진행
'''
```

```python
age_None = df[df['승객연령'].isin(['None'])].index
df.drop(age_None, inplace = True)
df
```

## 필요한 컬럼 생성

```python
df[['승차시각','하차시각']]
```

```python
df['승차년'] = df['승차시각'].astype(str).str[:4]
df['승차월'] = df['승차시각'].astype(str).str[4:6]
df['승차일'] = df['승차시각'].astype(str).str[6:8]
df['승차시'] = df['승차시각'].astype(str).str[8:10]
df['승차분'] = df['승차시각'].astype(str).str[10:12]
df['승차초'] = df['승차시각'].astype(str).str[12:]
```

```python
df = df.drop(['승차시각','하차시각'], axis=1)
df
```

## 데이터 저장
