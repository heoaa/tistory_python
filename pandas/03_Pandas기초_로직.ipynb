```python
code_master = pd.read_excel('./files2/sample_codemaster.xlsx')
i = 0
sample_total = pd.DataFrame()

# 기준년월 반복문(24번)
for yyyy in range(2019, 2021, 1) :
    for mm in range(1, 13, 1) :
        # print(str(yyyy) + '-' + str(mm).zfill(2))     # .zfill(n) : n자리수 지정(빈자리는 0으로 채움)
        
        # 파일 불러오기
        file_path = f'./files2/sample({i}).xlsx'
        sample = pd.read_excel(file_path, header=1, skipfooter=2, usecols='A:C')
        i += 1
        
        # 기준년월 변수 생성
        sample['기준년월'] = str(yyyy) + '-' + str(mm).zfill(2)
        
        # 국적코드 열기준 합침 (국적명 생성)
        sample_end = pd.merge(left = sample, right = code_master,
                               how = 'left',
                              left_on = '국적코드', right_on = '국적코드')
        
        sample_total = pd.concat([sample_total, sample_end], ignore_index=True)
        
print(len(sample_total))
sample_total
```

```python
ym_list = []
for yyyy in range(2019, 2021, 1) :
    for mm in range(1, 13, 1) :
        a = str(yyyy) + '-' + str(mm).zfill(2)     # .zfill(n) : n자리수 지정(빈자리는 0으로 채움)
        ym_list.append(a)

ym_list    
```

```python
code_master = pd.read_excel('./files2/sample_codemaster.xlsx')

for i in range(24) : 
    file_path = './files2/sample({}).xlsx'.format(i)
    sample = pd.read_excel(file_path, header=1, skipfooter=2, usecols='A:C')
    sample['기준년월'] = ym_list[i]
    
sample
```
