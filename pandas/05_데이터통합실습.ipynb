### 데이터 통합

```python
# 전체 116개 통합하기
# 통합에 사용할 데이터프레임 변수명 : df
# 전체 통합 후 조회
# 201001 ~ 201908
```

```python
df = pd.DataFrame()
ym = []
ym2 = []

for x in range(2010,2020) :
    for y in range(1, 13) :
        ym_a = str(x)+str(y).zfill(2)
        ym_b = str(x)+'-'+str(y).zfill(2)
        ym.append(ym_a)
        ym2.append(ym_b)
        if ym_a =='201908' :
            break


for k in range(0, len(ym)) :
    kto = pd.read_excel('./files/kto_{}.xlsx'.format(ym[k]), header=1, skipfooter=4, usecols='A:G')
    
    kto['기준년월'] = ym2[k]

    continents_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
    kto_country = kto[kto['국적'].isin(continents_list) == False]

    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 +\
                 ['오세아니아']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']

    kto_country = kto_country.reset_index(drop=True)

    kto_country['대륙'] = continents

    kto_country['관광객비율(%)'] = round((kto_country['관광'] /kto_country['계'])*100, 1)

    kto_country['전체비율(%)'] = round((kto_country['관광'] / kto_country['관광'].sum()) * 100, 1)

    df = pd.concat([df, kto_country], ignore_index=True)

df
```

```python
df = pd.DataFrame()

for y in range(2010, 2020, 1) :
    for m in range(1, 13, 1) :
        #예외처리(201909~)
        try :
        
            # 파일 처리
            file_name = str(y) + str(m).zfill(2)
            file_path = './files/kto_{}.xlsx'.format(file_name)
            kto = pd.read_excel(file_path, header=1, skipfooter=4, usecols='A:G')

            kto['기준년월'] = str(y)+'-'+str(m).zfill(2)


            continents_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
            kto_country = kto[kto['국적'].isin(continents_list) == False]

            continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 +\
                         ['오세아니아']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']
            
            kto_country = kto_country.reset_index(drop=True) ##

            kto_country['대륙'] = continents

            kto_country['관광객비율(%)'] = round((kto_country['관광'] /kto_country['계'])*100, 1)

            kto_country['전체비율(%)'] = round((kto_country['관광'] / kto_country['관광'].sum()) * 100, 1)

            df = pd.concat([df, kto_country], ignore_index=True)
            
        except :
            pass
df
```
