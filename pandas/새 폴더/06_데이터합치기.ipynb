```python
import pandas as pd
```

```python
# sample_1 데이터프레임 변수명으로 sample.xlsx 데이터 읽어들인 후 조회하기
file_path = 'C://STUDY//StudyPython_lee//day1_파이썬기초//files//sample_1.xlsx'
sample_1 = pd.read_excel(file_path, skiprows=[0,8,9], usecols=[0,1,2])
sample_1
```

```python
# sample_codemaster.xlsx 파일 읽은 후 조회
# 데이터프레임변수 : code_master
file_path = 'C://STUDY//StudyPython_lee//day1_파이썬기초//files//sample_codemaster.xlsx'
code_master = pd.read_excel(file_path)
code_master
```

#### merge

```python
# 열을 기준으로 합치기
```

```python
# 원본에 국적명 통합하기(merge)
pd.merge(sample_1,code_master)     # 컬럼명이 같은 경우 자동 매핑
```

```python
pd.merge(left = sample_1, right = code_master, 
         how = 'inner',         # 왼쪽을 기준으로 같은게 있는 경우만 매핑
         left_on = '국적코드', right_on = '국적코드')
```

```python
pd.merge(left = sample_1, right = code_master, 
         how = 'left',          # 왼쪽을 기준으로 있는대로 전부 가져와 매핑
         left_on = '국적코드', right_on = '국적코드')
```

```python
pd.merge(left = sample_1, right = code_master, 
         how = 'outer',          # 왼쪽 오른쪽 모두 있는대로 가져와 매핑
         left_on = '국적코드', right_on = '국적코드')
```

```python
# 데이터프레임변수명 : sample_2
# file : sample_2.xlsx

sample_2 = pd.read_excel('./files/sample_2.xlsx', header = 1, skipfooter = 2, usecols = [0,1,2])
sample_2
```

```python
sample_2['기준년월'] = '2019-12'
sample_2
```

#### concat

```python
# 행을 기준으로 합치기
```

```python
sample_total = pd.concat([sample_1, sample_2], 
                         ignore_index=True)      # 기존 데이터프레임 인덱스를 따라감
sample_total
```
