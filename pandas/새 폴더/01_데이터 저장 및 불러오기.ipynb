### 파일 불러와 저장하기

* skiprows = [] : 해당 행을 제외하고 불러옴
* usecols = [] or 'A:C' : 해당 컬럼만 불러옴
* skipfooter = a : 가장 아래 행에서 a개까지 제외한 후 불러옴

```python
# sample_1 데이터프레임 변수명으로 sample.xlsx 데이터 읽어들인 후 조회하기
file_path = 'C://STUDY//StudyPython_lee//day1_파이썬기초//files//sample_1.xlsx'
sample_1 = pd.read_excel(file_path, skiprows=[0,8,9], usecols=[0,1,2])
sample_1
```

```python
sample_1 = pd.read_excel(file_path, header = 1, skipfooter = 2, usecols = 'A:C')
sample_1
```

```python
# 저장하기
# file : sample.xlsx

sample_1.to_excel('./files/sample.xlsx', index = False)
```
