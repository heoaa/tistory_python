```python
# 필터링
# 성별이 남성인 데이터만 조회하기

sample_1[sample_1['성별'] == '남성']
```

```python
# 입국객수가 150000 이상인 데이터만 조회하기

sample_1[sample_1['입국객수'] >= 150000]
```

```python
# 성별이 여성이고, 입국객수가 150000이상인 데이터만 조회하기
a = sample_1[sample_1['성별'] == '여성']    # 변수처리 방식
a[a['입국객수'] >= 150000]
```

```python
sample_1[(sample_1['성별'] == '여성')&(sample_1['입국객수'] >= 150000)]    # 여러 조건일 경우 ()로 묶어줘야 함
```

```python
# 국적코드의 값이 A01 또는 A18인 데이터만 조회하기
sample_1[(sample_1['국적코드'] == 'A01')|(sample_1['국적코드'] == 'A18')]
```

```python
# 국적코드의 값이 A01 또는 A18인 데이터만 조회하기(isin 사용)
# isin() : 같은 컬럼내 여러개 값이 포함되어 있는지 확인할 때 사용(or과 유사)
sample_1[sample_1['국적코드'].isin(['A01','A18'])]
```

```python
sample_1[(sample_1['국적코드'].isin(['A01','A18'])) == False]
```

### 국적코드표 데이터 불러오기

```python
sample_1
```

```python
# sample_codemaster.xlsx 파일 읽은 후 조회
# 데이터프레임변수 : code_master

code_master = pd.read_excel('./files/sample_codemaster.xlsx')
code_master
```

#### merge

```python
# 열을 기준으로 합치기
```

```python
# 원본에 국적명 통합하기(merge)
pd.merge(sample_1,code_master)     # 컬럼명이 같은 경우 자동 매핑
```

```python
pd.merge(left = sample_1, right = code_master, 
         how = 'inner',         # 왼쪽을 기준으로 같은게 있는 경우만 매핑
         left_on = '국적코드', right_on = '국적코드')
```

```python
pd.merge(left = sample_1, right = code_master, 
         how = 'left',          # 왼쪽을 기준으로 있는대로 전부 가져와 매핑
         left_on = '국적코드', right_on = '국적코드')
```

```python
pd.merge(left = sample_1, right = code_master, 
         how = 'outer',          # 왼쪽 오른쪽 모두 있는대로 가져와 매핑
         left_on = '국적코드', right_on = '국적코드')
```

```python
# 데이터프레임변수명 : sample_2
# file : sample_2.xlsx

sample_2 = pd.read_excel('./files/sample_2.xlsx', header = 1, skipfooter = 2, usecols = [0,1,2])
sample_2
```

```python
sample_2['기준년월'] = '2019-12'
sample_2
```

#### concat

```python
# 행을 기준으로 합치기
```

```python
sample_total = pd.concat([sample_1, sample_2], 
                         ignore_index=True)      # 기존 데이터프레임 인덱스를 따라감
sample_total
```

```python
# 저장하기
# file : sample.xlsx

sample_total.to_excel('./files/sample.xlsx', index = False)
```
